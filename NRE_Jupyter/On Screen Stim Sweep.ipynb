```python
import pygame
import socket
import math


# input location of lab recorder and root of where you want to store data
#!start "" "C:\Users\lijun\OneDrive\Desktop\Jun_code\LabRecorder\LabRecorder.exe"
f="{root:C:\\Users\\lijun\\OneDrive\\Desktop\\Cognixion_data}"

participant = "Jun_June_26_Right_eye_Eye_closed"


# Initialize pygame
pygame.init()

# Set up display
screen_width = 1080
screen_height = 640
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Flashing Circle 7-13Hz')

# Define colors
BLACK = (0, 0, 0)

mod=1
WHITE = (round(mod*0), round(mod*255), round(mod*0))


# Circle settings
circle_radius = 50
circle_position = (screen_width // 2, screen_height // 2)

# Set up font
font = pygame.font.Font(None, 74)

#Initiazing
message = str(1)
Freq=1
circle_visible = True


# Main game loop
running = True
last_flash_time = pygame.time.get_ticks()  # Get the initial time in milliseconds





#########################
# Input variable / parameter

Starting_frequency = 7 #Hz
Ending_frequency= 13 #Hz
desired_duration = 15  # in second
increment = 1   # in Hz
waittime=  2  # in second
Participant = "{participant: %s}" %participant

# Change to directory for your data


#user_input = input("Please enter participant name: ")
#Participant = "{participant: %s}" %user_input
#print(Participant)
#freq= f"{{session: {Freq}}}"
#task="task: Default"
#########################


soc = socket.create_connection(("localhost", 22345))
#soc.sendall(
soc.sendall(b"select cognixion_raw_eeg (localhost)\n")



while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            soc.sendall(b"stop\n")
            pygame.quit()
        
    
    for x in range(0,int((Ending_frequency-Starting_frequency)/increment)+1):
        Freq=Starting_frequency+ x * increment
        if running == False:
            break

        #print(Freq)
        soc.sendall(b"stop\n")
        pygame.time.wait(1500)
        #config= "filename " + f + Participant + "{session: " + str(Freq) +"}\n"
        template = "{template:%s\\%s.xdf}" %(participant, Freq)
        config= "filename " + template + f + Participant + "{session: " + str(Freq) +"}\n"
        #print(config)
        soc.sendall(b"Update\n")
        soc.sendall(config.encode())
        
        soc.sendall(b"start\n")
        
        screen.fill(BLACK)
        # Render the message text
        message = "Waiting for Labrecorder"
        text = font.render(message, True, WHITE)
        text_rect = text.get_rect(center=(screen_width // 2, screen_height // 4))  # Recalculate position in case of changes
        screen.blit(text, text_rect)

        pygame.time.wait(2500)


        
        #timestamp_start=pygame.time.get_ticks()
        initial_time= current_time = pygame.time.get_ticks()  

        # For loop range chosen to be an arbitrary value higher than trial duration
        # Loop break out when time reach threshold instead of finishing looping
        for timer in range(initial_time, initial_time+desired_duration*3000):
            
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    # Quit pygame
                    soc.sendall(b"stop\n")
                    pygame.quit()

            current_time = pygame.time.get_ticks()
            if(current_time-initial_time >= desired_duration*1000):
                break
        

            last_flash_time = current_time  # Reset last flash time
                
            ##########################################
            # Update Frequecy on display
            # Fill the screen with black
            screen.fill(BLACK)
            # Render the message text
            message = str(Freq)
            text = font.render(message, True, WHITE)
            text_rect = text.get_rect(center=(screen_width // 2, screen_height // 4))  # Recalculate position in case of changes
            screen.blit(text, text_rect)
            ##########################################
        
            # Draw the circle
            sine_value = (math.sin(current_time * Freq * 2 * math.pi / 1000)+1)/2   # sine wave oscillating between 0 and 1
            #sine_value_2=(math.sin(current_time * (Freq+2) * 2 * math.pi / 1000)+1)/2 
            #combined= (sine_value + sine_value_2)/2
            circle_brightness = (int(255 * sine_value),int(255 * sine_value),int(255 * sine_value))  # Adjust brightness based on sine wave
            #circle_brightness = (int(255 * combined),int(255 * combined),int(255 * combined))
            #circle_brightness = (int(255 * 0),int(255 * sine_value),int(255 * 0))
            pygame.draw.circle(screen, circle_brightness, circle_position, circle_radius)
            #pygame.draw.rect(screen, circle_brightness, pygame.Rect(circle_position[0], circle_position[1], circle_radius*2, circle_radius*2))
            
            
            #sine_value = (math.sin(current_time * Freq * 2 * math.pi / 1000)+1)/2   # sine wave oscillating between 0 and 1
            #circle_black = (0,0,0)
            #pygame.draw.circle(screen, circle_black, circle_position, 0.2*circle_radius)
            
            # Update the display
            pygame.display.flip()
            
            # Control the clock rate
            pygame.time.Clock().tick(2000)
            
        soc.sendall(b"stop\n")
        pygame.time.wait(2000)
        

        screen.fill(BLACK)

        ###################################
        # Update Rest Message
        message = "Rest for " + str(waittime) + " Second"
        text = font.render(message, True, WHITE)
        text_rect = text.get_rect(center=(screen_width // 2, screen_height // 4))  # Recalculate position in case of changes
        screen.blit(text, text_rect)
        pygame.display.flip()
        ###################################
        #modify the rest time in millisecond
        pygame.time.wait(int(waittime*1000))

# Quit pygame
    soc.sendall(b"stop\n")
    pygame.quit()
```
