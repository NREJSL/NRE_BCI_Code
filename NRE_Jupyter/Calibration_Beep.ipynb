```python
import pygame
import socket
import math
# Initialize the mixer


# input location of lab recorder and root of where you want to store data
#!start "" "C:\Users\lijun\OneDrive\Desktop\Jun_code\LabRecorder\LabRecorder.exe"
f="{root:C:\\Users\\lijun\\OneDrive\\Desktop\\Cognixion_data}"

participant = "Jun_June_26"

# Initialize pygame
pygame.init()

# Set up display
screen_width = 640
screen_height = 380
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Flashing Circle 7-13Hz')

# Define colors
BLACK = (0, 0, 0)

mod=1
WHITE = (round(mod*0), round(mod*255), round(mod*0))


# Circle settings
circle_radius = 50
circle_position = (screen_width // 2, screen_height // 2)

# Set up font
font = pygame.font.Font(None, 74)

#Initiazing
message = str(1)
Freq=1
circle_visible = True


# Main game loop
running = True
last_flash_time = pygame.time.get_ticks()  # Get the initial time in milliseconds

#########################
# Input variable / parameter

Starting_frequency = 7 #Hz
Ending_frequency= 14 #Hz
desired_duration = 20  # in second
increment = 1  # in Hz
waittime=  2  # in second
Participant = "{participant: %s}" %participant

soc = socket.create_connection(("localhost", 22345))
#soc.sendall(
soc.sendall(b"select cognixion_raw_eeg (localhost)\n")
pygame.mixer.init()

# Load and play the MP3 file
pygame.mixer.music.load("beep.mp3")



while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            soc.sendall(b"stop\n")
            pygame.quit()
        
    
    for x in range(0,int((Ending_frequency-Starting_frequency)/increment)+1):
        Freq=Starting_frequency+ x * increment
        if running == False:
            break

        #print(Freq)
        soc.sendall(b"stop\n")
        pygame.time.wait(1500)
        #config= "filename " + f + Participant + "{session: " + str(Freq) +"}\n"
        template = "{template:%s\\%s.xdf}" %(participant, Freq)
        config= "filename " + template + f + Participant + "{session: " + str(Freq) +"}\n"
        #print(config)
        soc.sendall(b"Update\n")
        soc.sendall(config.encode())
        
        soc.sendall(b"start\n")
        screen.fill(BLACK)
        # Render the message text
        message = "Waiting for Labrecorder"
        #

        pygame.time.wait(1500)


        
        pygame.mixer.music.play()
        pygame.time.wait(250)
        
        # Keep the script running until the music stops
        while pygame.mixer.music.get_busy():
            pygame.time.Clock().tick(10)

        pygame.time.wait(desired_duration*1000)
       
        soc.sendall(b"stop\n")
        pygame.time.wait(1000)
        
        pygame.mixer.music.play()
        pygame.time.wait(250)
        pygame.mixer.music.play()
        pygame.time.wait(250)
        pygame.mixer.music.play()
      

        screen.fill(BLACK)

        ###################################
        # Update Rest Message
        message = "Rest for " + str(waittime) + " Second"
        text = font.render(message, True, WHITE)
        text_rect = text.get_rect(center=(screen_width // 2, screen_height // 4))  # Recalculate position in case of changes
        screen.blit(text, text_rect)
        pygame.display.flip()
        ###################################
        #modify the rest time in millisecond
        pygame.time.wait(int(waittime*1000))
        screen.fill(BLACK)

# Quit pygame
    soc.sendall(b"stop\n")
    pygame.quit()
```
